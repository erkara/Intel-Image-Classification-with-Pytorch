First of all, lets look at some of the images. We write small commment for the helper function.

```python
def GetImgLabels(parentdir):
    #returns images paths with the labels as list of tuples
    # [(img1_path,img1_label),...]
    d = {}
    for folder in os.listdir(parentdir):
        path = os.path.join(parentdir,folder)
        
        for img in os.listdir(path):
            
            img_path = os.path.join(path,img)
            img_label = folder
            
            d[img_path] = img_label  
    keys = list(d.keys())   
    random.shuffle(keys)
    l = [(key, d[key]) for key in keys]
    return l
```

```python
def ViewSamples(n,parentdir):
    # show nxn random samples from parentdir
    imgs_labels = GetImgLabels(parentdir)
    f,ax = plt.subplots(n,n) 
    f.subplots_adjust(0,0,3,3)#(left,bottom,vertical_distance b/w columns)
    for i in range(0,n,1):
        for j in range(0,n,1):
            rnd_number = random.randint(0,len(imgs_labels))
            img_path,img_label = imgs_labels[rnd_number]
            image = cv2.cvtColor(cv2.imread(img_path), cv2.COLOR_BGR2RGB)
            ax[i,j].imshow(image)
            ax[i,j].set_title(img_label)
            ax[i,j].axis('off')

ViewSamples(4,test_dir)
```

We define three funtions

- process_image takes a row image(numpy array), resizes to 150x150, normalizes using ImageNet statistics and returns as a torch tensor. In Pytorch, we feed the model with the matrices of the form (B,C,W,H) where B is the batch size, C is the color channel(feature maps) and W,H are the image width and hight, respectively. We need to reshape the image at the end since OpenCV has (W,H,C) structure. Also BRG to RGB conversion is required to display the true image in matplotlib enviroment.
- imshow takes a torch tensor and tranforms back to numpy array and displays as an image.
- predimgpaths returns all the prediction image paths.


```python
def process_image(imagepath):
    # returns the normalized image in torch form.
    img_size = 150
    img = cv2.resize(cv2.imread(imagepath),(img_size,img_size))
    image = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    
    #normalize the image.
    np_image = np.array(image)/255
    mean = np.array([0.485, 0.456, 0.406])
    std = np.array([0.229, 0.224, 0.225])
    np_image = (np_image - mean) / std
    
    #torch expects first dim to be color channel.
    np_image = np_image.transpose((2, 0, 1))
    
    return np_image
```

```python
def imshow(torch_img):
    # takes a torch image and displays by converting back.
    #convert to (W,H,Channel)
    image = torch_img.transpose((1, 2, 0))
    mean = np.array([0.485, 0.456, 0.406])
    std = np.array([0.229, 0.224, 0.225])
    #revert normalization
    image = std * image + mean
    plt.imshow(image)
    #plt.show()
```

```python
def predimgpaths(pred_dir):
    #return all pred file paths
    pred_path_list = []
    for files in os.listdir(raw_dir):
        pred_path_list.append(os.path.join(raw_dir,files))
    return pred_path_list
```

We load the model we trained in the previous section.

```python
def load_checkpoint(filepath):
    #load the model and return in eval() mode.
    checkpoint = torch.load(filepath)
    model = checkpoint['arch']
    model.class_to_idx = checkpoint['class_to_idx']
    model.load_state_dict(checkpoint['state_dict'])
    for parameter in model.parameters():
        parameter.requires_grad = False

    model.eval()
    return model

model = load_checkpoint('/home/erdidenizli88/MyDnnWorksMachine2/IntelImage/IntelModel32.pth')
```

We now define the prediction function for an input image. It displays top-k probability with its classes predicted by the model.

```python
def predict(image_path, model, topk=4):
    ''' Predict the class (or classes) of an image using a trained deep learning model.
    '''
    model.to('cuda')
    model.eval()
    
    image = process_image(image_path)
    
    # Convert image to PyTorch cuda-tensor
    image = torch.from_numpy(image).type(torch.cuda.FloatTensor)
    
    
    # torch expects (Batch,Color,W,H). This converts to (1,3,150,150)
    image = image.unsqueeze(0)
    
    #forward pass
    output = model(image)
    
    #evaluate the probabilities
    probabilities = torch.exp(output)
    
    # Probabilities and the indices of the corresponding classes.
    top_probabilities, top_indices = probabilities.topk(topk)
    
    # Convert to lists. Note that output carries requires_grad=true, so detach it then convert.
    top_probabilities = top_probabilities.detach().type(torch.FloatTensor).numpy().tolist()[0] 
    top_indices = top_indices.detach().type(torch.FloatTensor).numpy().tolist()[0] 
    
    # Convert topk_indices to the actual class labels using class_to_idx
    # Invert the dictionary to get a mapping from index to class.
    
    idx_to_class = {value: key for key, value in train_set.class_to_idx.items()}
    #print(idx_to_class)
    
    top_classes = [idx_to_class[index] for index in top_indices]
    
    return top_probabilities, top_classes
```

ShowPredictions displays the results for images randomly drawn from pred_dir

```python
def ShowPredictions(n,pred_dir):
    pred_path_list = predimgpaths(pred_dir)
    plt.subplots(n,2,figsize=(12, 4*n)) 
    plt.subplots_adjust(0,0,3,3,wspace = 0.5)
    for i in range(1,n+1):
        rnd_number = random.randint(0,len(pred_path_list))
        img_path = pred_path_list[rnd_number]
        np_image = process_image(img_path)
        plt.subplot(n,2,2*i-1)
        imshow(np_image)
        probs, classes = predict(img_path, model)
        plt.subplot(n,2,2*i)
        plt.barh(classes, probs,align='center')
        plt.tight_layout()
    plt.show()
    
ShowPredictions(5,raw_dir)
```

Lets create the confusion matrix. We will import a scikit learn module 'confusion_matrix'. To display the matrix, lets use the 'seaborn' library which is built on the top of matplotlib Note that we define GetConfusionMatrix with @torch.no_grad() decorator since we only need to evaluate the results. This will speed up the computation.

```python
test_loader = DataLoader(test_set,batch_size=60)
```

```python
@torch.no_grad()
def GetConfusionMatrix(model,loader):
    model.eval()
    model.to(device)
    all_preds  = torch.cuda.FloatTensor([])
    #compute all predictions
    for batch in loader:
        images,labels = batch
        images,labels = images.to(device),labels.to(device)
        preds = model(images)
        all_preds = torch.cat(
            (all_preds,preds),
            dim=0
        )
    #get the predicted labels
    preds = all_preds.argmax(dim=1)
    #get the correct labels
    targets = torch.tensor(loader.dataset.targets,dtype=torch.long)
    #create confusion matrix
    cm = confusion_matrix(preds.cpu().numpy(), targets)
    return cm

cm = GetConfusionMatrix(model,test_loader)
cm
```

We now introduce several ways to measure the quality of the model. Accuracy is not always the best way to report the model performance. We drive new scores from the confusion matrix. Based on the application, we should make sure some of these scores are not considirably lower than the others. 

Overal Accuracy

```python
# Be cautius, not always the best way
OveralAcc = np.sum(np.diag(cm))/np.sum(cm)
print('OverallAccuracy', OveralAcc)
```

This results actually confirms our findings in the previous notebook.

RECALL

Recal tells us the ability of algoritm to classify that class. 0.905 recall means out of 100 bulding images 90 were correctly predicted. Results below show that the algorithm struggles to correctly classify the glacier and moutain images.

```python
Recall = np.diag(cm)/np.sum(cm,axis = 1)
print('Recals:', Recall)
MeanRecall = np.mean(Recall)
print('Mean_Recall:', MeanRecall)
```

PRECISION

Precision tells us the ability of an alg not to label anything as "building". 0.94 precison means only 6 out of 100 images were falsely predicted as building.

```python
Prec = np.diag(cm)/np.sum(cm,axis = 0)
print('Precisions:', Prec)
MeanPrecision = np.mean(Prec)
print('Mean_Precision:', MeanPrecision)
```

F-SCORE

This is the harmonic average of recall recall and precision. It is a realiable measure putting equal weigths on precison and recall.

```python
F_Score = 2/(1/Recall + 1/Prec)
print('F-Scores:', F_Score)
MeanF_Score = np.mean(F_Score)
print('Mean_FScore:', MeanF_Score)
```
