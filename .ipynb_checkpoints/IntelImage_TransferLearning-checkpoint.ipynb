```python
import numpy as np
from matplotlib import pyplot as plt
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
import torchvision
from torchvision import datasets, transforms, models
from torch.utils.data import DataLoader
from torch.utils.data import sampler
from torch.utils.data import SubsetRandomSampler
from torch._six import int_classes as _int_classes
from tensorboardX import SummaryWriter
from IPython.display import display, clear_output
import pandas as pd
import seaborn as sn
import time
from itertools import product
from collections import OrderedDict
import datetime
import warnings
import cv2
import os
import random
from sklearn.metrics import confusion_matrix
warnings.filterwarnings('ignore')
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 20)
pd.set_option('precision', 4)
```

```python
data_dir='~/MyDnnLibrary/Data/IntelImage'
train_dir = data_dir + '/train'
test_dir = data_dir + '/test'
raw_dir = data_dir + '/raw'
```

```python
# Define transforms for the training and testing.
training_transforms = transforms.Compose([transforms.Resize(size = (150,150)),
                                          transforms.RandomPerspective(),
                                          transforms.RandomVerticalFlip(),
                                          transforms.RandomHorizontalFlip(),
                                          transforms.ToTensor(),
                                          transforms.Normalize([0.485, 0.456, 0.406], 
                                                               [0.229, 0.224, 0.225])]
                                            )

test_transforms = transforms.Compose([    transforms.Resize(size = (150,150)),
                                          transforms.ToTensor(),
                                          transforms.Normalize([0.485, 0.456, 0.406], 
                                                               [0.229, 0.224, 0.225])]
                                    )

train_set = datasets.ImageFolder(train_dir, transform=training_transforms)
test_set = datasets.ImageFolder(test_dir, transform=test_transforms)
```

Training a large scale model on cpu is almost impossible in terms of computation time, which makes the use of GPU essential We will train and test the model on Google Compute Engine with NVIDIA Tesla P100. 

```python
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
```

We will be using vgg16 as a pretrained model. We should note that the original VGG arhitecture was trained on 224x224 images but we have 150x150 images. This means the parameters of the model were tuned using 224x224 images. However this model works without error for any input size since it deploys Adaptive Average Pooling before the linear layer. With this special filter, any input size is scaled down to 7x7 before the fully connected layer. This allows us to fine-tune vgg based on our image sizes. We will follow this approach here. Another approach, which is very common, is to resize the input image to 224x224 and then feed to the model. However such scaling may introduce a consdirable noise and distortion in the image and also significantly increase the number of paramaters, thus training time. In this notebook, we will follow the former approach and stay on 150x150 and try to fine tune the model with this setting. We will then create another notebook with the later approach. 
We can look at the architecture:

```python
model = models.vgg16(pretrained=True)
model
```

We first freeze the model. By this way no backpropagation across the frozen layers is performed during training. In other words, we dont train those layers.

```python
for parameter in model.parameters():
    parameter.requires_grad = False
```

Notice that out_features=1000 in the ouput layer. This is because VGG16 was trained on 1000 different classes. We will customize the model based on our problem. First of all, we would not want to change the early layers since it is already well tuned. Thus we will only modify the second sequential layer. The easiest way to achieve this is to replace the output layer with the number of classes in our data, which is 6. This can be done with

model.classifier[6]= nn.Linear(4096, 6)              
                      
In default, all new layers are trainable. In this case, we can train the model with a little effort even though there is no way to predict the performance. This model would consist of 4096*6+6=24582 traninable parameters. However we will add two fully connected layers with 256 neurons each with a dropout layer and the output layer will be LogSoftmax which is the natural log version of the usual softmax layer. 

```python
model.classifier[6] = nn.Sequential(
                      nn.Linear(4096, 256), 
                      nn.ReLU(), 
                      nn.Dropout(0.3),
                      nn.Linear(256, 6),                   
                      nn.LogSoftmax(dim=1))
```

Now we can have a look at the details of the modified model. Notice that 1,050,374 is the number of parameters we added and they are trainable but rest will remain unchanged over the course of the training.

```python
summary(model.to('cuda'),input_size=(3, 150, 150))
```

```python
def TrainTime(run_count,epoch,num_epochs, epoc_duration, runtotal):
    #get some time statistics in the training.
    total_min = num_epochs*runtotal*epoc_duration/60
    total_hour = total_min/60
    rem_min = ((runtotal-run_count+1)*num_epochs - epoch)*epoc_duration/60
    rem_hour = rem_min/60
    print("Aproximate Training time: {m:2.2f} mins / {h:2.2f} hours \
          Approximate Remaining Time: {rm:2.2f} mins / {rh:2.2f} hours".format(m=total_min, h=total_hour,rm=rem_min,rh=rem_hour))
    
def GetFilename(name='TransferNew_'):
    #we use this function to create uniqe filenames to label various statistics.
    now = datetime.datetime.now()
    now = str(now.strftime("%Y-%m-%d %H:%M"))
    now = name+now.replace(' ', '_time_').replace(':', '-')
    return now

def GetBestAcc(filename,n):
    #get the best n test accuracy of the cuurent experiment.
    df = pd.read_csv(filename)
    sort_by_test_accuracy = df.sort_values('TestAccuracy',ascending=False)
    display(sort_by_test_accuracy.head(n=n))
    
def GetBestStats(FinalResults,HyperParameters):
    #tags for the final models for further inference.
    df = pd.read_csv(FinalResults,float_precision='0.2f')
    beststats = {'max_epoch_num':      df['epoch'].max(),
            'BestTrainLoss':      df['TrainLoss'].min(),
            'BestTrainAccuracy':  df['TrainAccuracy'].max(),
            'BestTestLoss':       df['TestLoss'].min(),
            'BestTestAccuracy':   df['TestAccuracy'].max(),
            'lr':                 HyperParameters['lr'][0],
            'batch_size':         HyperParameters['batch_size'][0],
            'weight_decay':       HyperParameters['weight_decay'][0]
           }
    return beststats
```

```python
def save_checkpoint(model,name,final_results):
    # save the model with the given info. results must be provided as csv.
    beststats = GetBestStats(final_results,HyperParameters)
    
    model.class_to_idx = train_set.class_to_idx
    checkpoint = {'arch': model,
                 'class_to_idx': model.class_to_idx,
                 'state_dict': model.state_dict(),
                }
    checkpoint.update({'BestScores':beststats})

    torch.save(checkpoint, name+'.pth')
```

```python
def load_checkpoint(filepath):
    #load the model and return in eval() mode.
    checkpoint = torch.load(filepath)
    model = checkpoint['arch']
    model.class_to_idx = checkpoint['class_to_idx']
    model.load_state_dict(checkpoint['state_dict'])
    model.best_scores = checkpoint['BestScores']
    for parameter in model.parameters():
        parameter.requires_grad = False

    model.eval()
    return model
```

This is our main function to start and output all statistics. All items are self explanatory but we note a couple of points here:
- GPU communication is performed through CUDA. Thus we sent all tensors to cuda to perform the calcuations on the GPU.
- We free the GPU cache after every batch with torch.cuda.empty_cache(), otherwise we may run into "CUDA out of memory" error. Altough this is not an issue for a small scale problems, GPU can easily run out of memory due to the large workload of forward and backward passes.
- We consider learning rate, batch size and weight_decay as a hyperparamater family. However we will keep weight decay as a safe guard to overfitting. In fact we will force our model to overfit, it it does, we will turn on weight decay to balance overfitting. We did not encounter overfitting in this model so kept it 0. 
- We also define a max_acc = 0.90.  The following code updates the max_acc for every higher score achieved after each epoch and assign the model's current state to opt_model. So the algoritm returns the model which has the highest testing accuracy. 
- We commented the lines concerning Tensorboard since we already analyze the output with Pandas but we keep the lines in the code so that we can monitor the outputs on tensorbaord as well. 

```python
def StartTrain(num_epochs, runtotal):
    run_count = 0
    df = pd.DataFrame(columns=['run', 'epoch', 'TrainLoss','TrainAccuracy','TestLost','TestAccuracy',
                               'lr','batch_size','weight_decay','epoch_duration'])
    max_acc = 0.92
    for lr, batch_size, weight_decay in product(*param_values):
        run_count += 1
        
        torch.cuda.empty_cache()
        
        model.to(device)
        
        train_loader = DataLoader(train_set,batch_size=batch_size,shuffle=True )
        test_loader = DataLoader(test_set,batch_size=batch_size)

        optimizer = optim.Adam(model.classifier.parameters(), lr=lr,weight_decay=weight_decay,)
        
        
        # This works well with logsoftmax.
        criterion = nn.NLLLoss()

        # Crete tensorboard runs for visualize
        comment = f'batch_size={batch_size} lr={lr} weight_decay={weight_decay}' 
        tb = SummaryWriter(log_dir = 'runs/'+ GetFilename(),comment=comment)
        
        #sent some images if needed. 
#         images,labels = next(iter(train_loader))
#         images = images.to(device)
#         labels = labels.to(device)
#         grid = torchvision.utils.make_grid(images)
#         tb.add_image('images',grid)
#         tb.add_graph(model,images)

        for epoch in range(num_epochs):
            epoch_start = time.time()
            # set to train mode
            model.train()

            # compute the train loss and correct predictions per epoch
            train_loss = 0.0
            total_train_correct = 0

            # compute the test loss and correct predictions per epoch
            test_loss = 0.0
            total_test_correct = 0

            for i,(images, labels) in enumerate(train_loader):

                images = images.to(device)
                labels = labels.to(device)

                #In default, pytorch accumulates the gradient, zero out before each backpropagation.
                optimizer.zero_grad()

                # forward pass
                outputs = model(images)

                # compute the loss
                loss = criterion(outputs, labels)

                # backpropagate the gradients
                loss.backward()

                # update the parameters
                optimizer.step()

                # compute the total loss for the batch and add it to train_loss
                train_loss += loss.item() * images.size(0)

                # compute the predictions
                predictions = outputs.argmax(dim=1)

                # boolen compare the predictions vs labels and create the correct count tensor
                correct_counts = predictions.eq(labels)

                # convert the cuda tensor to FloatTensor and compute the mean for the batch
                acc = torch.mean(correct_counts.type(torch.FloatTensor)) 

                # accumulate the total train correct and update over the batches.
                total_train_correct += acc.item() * images.size(0) 
                #torch.cuda.empty_cache()


            # For testing, no need to track gradients.     
            with torch.no_grad():

                # set to evaluation mode
                model.eval()


                for j,(images, labels) in enumerate(test_loader):
                    images = images.to(device)
                    labels = labels.to(device)

                     # forward pass
                    outputs = model(images)

                    # compute the loss
                    loss = criterion(outputs, labels)

                    # compute the loss for the batch and add it to total test_loss
                    test_loss += loss.item() * images.size(0)

                    # compute the predictions
                    predictions = outputs.argmax(dim=1)

                    # boolen compare the predictions vs labels and create the correct count tensor
                    correct_counts = predictions.eq(labels)

                    # convert the cuda tensor to FloatTensor and compute the mean for the batch
                    acc = torch.mean(correct_counts.type(torch.FloatTensor))

                    # accumulate the total train correct and update over the batches.
                    total_test_correct += acc.item() * images.size(0)

                    
                    torch.cuda.empty_cache()

            # training loss and training accuracy
            avg_train_loss = train_loss/len(train_set) 
            avg_train_acc = total_train_correct/float(len(train_set))

            # test loss and test accuracy
            avg_test_loss = test_loss/len(test_set) 
            avg_test_acc = total_test_correct/float(len(test_set))
            

            epoch_end = time.time()

           #may want to check the computing performance.
            epoch_duration = epoch_end-epoch_start

            #tensorboard scaler variables
            tb.add_scalar('Traing_Loss', avg_train_loss,epoch)
            tb.add_scalar('Train_Accuracy',avg_train_acc,epoch)
            tb.add_scalar('Test_Loss',avg_test_loss,epoch)
            tb.add_scalar('Test_Accuracy',avg_test_acc,epoch)

            #print progress
            clear_output(wait=True)
            TrainTime(run_count,epoch,num_epochs, epoch_duration, runtotal)
            df.loc[int(num_epochs*(run_count-1)+epoch)] = [run_count, epoch+1, avg_train_loss ,avg_train_acc,
                                                     avg_test_loss,avg_test_acc,
                                               lr,batch_size,weight_decay,epoch_duration]
            df.to_csv(filename,index=False)
            display(df)
            
            
            if avg_test_acc > max_acc:
                opt_model = model
                max_acc = avg_test_acc
        
    return opt_model 
```

First of all, we dont generally encounter such graphs where the model performs better on testing data than trainig. However this pattern does not pose any threat to the model. Training images undergo three random transformations at each epoch so the model may struggle to capture the underlying data. In contrary, test data has a low variance so we are constantly getting better results. If we train more epochs, we can expect this patter to change. As we can see from the graph, there is no sharp divergence between the train-test acccuracy and loss. Most importantly, training loss is decreasing while test accuracy is getting better. At the first place, this means we are not overfitting. Moreover the model keeps learning but slowly. This indicates that it has a potential to learn with these configuration. However it is more difficult to improve the model after this point. That very slow loss in the training may be slowdown problem  or most of the neurons are near saturation as mentioned in the report.

We have many options (probably infintely many) at this stage. Since we cant work on training foreover, we will try to make a couple of adjustments and not to fall back ~%90 we achieved above. Let's increase the number of neorons in the linear layers, adding a batch normalization and decreasing the effect of dropout from 50% to 10% and use a smaller lr=3e-5 with a smaller batch size=32. We actually force the model overfitting. If we observe overfitting , we can then go back to 50% droput or even turn on weight decay but intiuitevely we can say we will not severely overfit. Lastly we will train more epochs to see how the model will respond. More epoch may enable the model to move deeper through the minimum.

```python
#second trial with batch normalizationn with more neurons, less dropout.
for parameter in model.parameters():
    parameter.requires_grad = False

classifier = nn.Sequential(OrderedDict([('fc1', nn.Linear(25088, 2000)),
                                        ('batch1', nn.BatchNorm1d(2000)),
                                        ('relu1', nn.ReLU()),
                                        ('drop', nn.Dropout(p=0.1)),
                                        ('fc2', nn.Linear(2000, 6)),
                                        ('output', nn.LogSoftmax(dim=1))]))

model.classifier = classifier
```
