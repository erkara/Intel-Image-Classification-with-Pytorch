```python
#confusion matrix.
cm  = np.array([
       [413,   1,   1,   2,   2,  37],
       [  0, 470,   2,   0,   3,   0],
       [  0,   1, 472,  55,   8,   0],
       [  0,   2,  56, 457,   3,   2],
       [  3,   0,  19,   8, 493,   0],
       [ 21,   0,   3,   3,   1, 462]])
```

Overal Accuracy

```python
# Be cautius, not always the best way
OveralAcc = np.sum(np.diag(cm))/np.sum(cm)
print('OverallAccuracy', OveralAcc)
```

RECALL 

```python
#Recal tells us the ability of algoritm to classify that class. 0.90 recal means out of 100 bulding images 90 
#were correctly predicted.
Recall = np.diag(cm)/np.sum(cm,axis = 1)
print('Recals:', Recall)
MeanRecall = np.mean(Recall)
print('Mean_Recall:', MeanRecall)
```

PRECISION

```python
# Precision tells us the ability of an alg not to label anything as "building". 0.94 precison means only 
#6 out of 100 images were falsely predicted as building.
Prec = np.diag(cm)/np.sum(cm,axis = 0)
print('Precisions:', Prec)
MeanPrecision = np.mean(Prec)
print('Mean_Precision:', MeanPrecision)
```

F-SCORE

```python
#This is the harmonic average of recall recall and precision. It is the most reliable measure.
F_Score = 2/(1/Recall + 1/Prec)
print('F-Scores:', F_Score)
MeanF_Score = np.mean(F_Score)
print('Mean_FScore:', MeanF_Score)
```
